## Format primary data type

```python
# Using f-string
```

```python
number_1: int = 1
print(f"{number_1 = }")

def add(*args) -> None:
    print(f"{sum(args) = }")
add(1,2,3,4)
```

```python
# Add some seperator for string
print(f"{' Hello ':=^30}") # 30 = ' Hello ' (7) + '=' (23)
print(f"{' Hello ':*^30}")
print(f"{'Hello ':=<30}")
print(f"{' Hello':=>30}")

string_1: str = ' Hello world, I am PAN. This is string for beautify. '
print(f"{string_1:=^70}")
```

```python
# Using Template for mutipleline string
```

```python
from string import Template

template_1: Template = Template("""Hello, I'm ${name}.
I'm ${age} years old.
""")
print(template_1.safe_substitute(name="Bob", age=25))
```

## Format object

```python
# Using json.dumps + print
import json

dict_1: dict = {
    "name": "A",
    "age": 1,
    "refered": {
        "name": "B",
        "age": 2
    }
}
print(json.dumps(dict_1, indent=2))

list_1: list = [
    {
        "id": 1,
        "name": "A"
    },
    {
        "id": 2,
        "name": "B"
    }
]
print(json.dumps(list_1, indent=2))
```

```python
from requests import Response, get
import json

res = get("https://realpython.com")
print("[{}] Response header:\n{}".format(__name__, json.dumps(dict(res.headers), indent=4)))
```

```python
from pprint import pp

pp(dict(res.headers), depth=100, width=40, sort_dicts=False)
```

```python
from pprint import PrettyPrinter
cusprint = PrettyPrinter(
    # dict(res.headers),
    width=79,
    depth=3,
    compact=True,
    sort_dicts=False,
    # underscore_numbers=True
    )
cusprint.pprint(dict(res.headers))
```
