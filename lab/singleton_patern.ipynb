```python
class SingletonA:
    _instance = None  # Class variable to hold the single instance

    def __new__(cls, *args, **kwargs):
        # 1. Check if an instance already exists
        if cls._instance is None:
            # 2. If not, call the base class's __new__ to create it
            cls._instance = super().__new__(cls)
            print("SingletonA instance created.")
        # 3. Return the single instance (either newly created or existing)
        return cls._instance

    def __init__(self, name=None):
        # We can still run initialization logic, but it will run every time
        # you try to create the instance (though __new__ ensures only one object)
        if not hasattr(self, '_initialized'):
            self.name = name if name else "Default Singleton"
            self._initialized = True
            print(f"SingletonA initialized with name: {self.name}")
        else:
            print("SingletonA already initialized. Skipping init logic.")

# --- Example Usage ---
# First call creates and initializes the instance
s1 = SingletonA(name="Logger1")
print(f"s1 name: {s1.name}")

# Second call returns the existing instance; __new__ runs, but doesn't create a new object.
s2 = SingletonA(name="Logger2") 
print(f"s2 name: {s2.name}")

# Check if both variables point to the same object
print(f"Are s1 and s2 the same object? {s1 is s2}")
```

```python
class SingletonMeta(type):
    """
    A metaclass that implements the Singleton pattern.
    """
    _instances = {}

    def __call__(cls, *args, **kwargs):
        # If the class (cls) is not in our instances dictionary, create it
        if cls not in cls._instances:
            # This calls the class's __new__ and then __init__ as normal.
            instance = super().__call__(*args, **kwargs)
            cls._instances[cls] = instance
        
        # Always return the stored instance
        return cls._instances[cls]

class DatabaseConnection(metaclass=SingletonMeta):
    """
    This class will be a Singleton because it uses SingletonMeta.
    """
    def __init__(self, db_name):
        if not hasattr(self, 'connection_id'):
            self.db_name = db_name
            # Simulating a resource heavy connection setup
            self.connection_id = hash(db_name)
            print(f"Connection created for: {self.db_name} (ID: {self.connection_id})")
        else:
            print(f"Connection already exists. Returning existing connection ID: {self.connection_id}")

# --- Example Usage ---
db1 = DatabaseConnection("production_db")
# The __init__ method for db2 will see the existing connection_id
db2 = DatabaseConnection("test_db") 

print(f"DB1 Name: {db1.db_name}, ID: {db1.connection_id}")
print(f"DB2 Name: {db2.db_name}, ID: {db2.connection_id}")

# Check for identity
print(f"Are db1 and db2 the same object? {db1 is db2}")
```
