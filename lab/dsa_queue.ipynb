### Basic using list

```python
class Queue:
    def __init__(self) -> None:
        self.__items: list = []
    
    def enqueue(self, item: any) -> None:
        self.__items.append(item)
    
    def dequeue(self) -> any:
        if self.is_empty():
            raise Exception("Queue is empty")
        return self.__items.pop(0)

    def peek(self) -> any:
        if self.is_empty():
            raise Exception("Queue is empty")
        return self.__items[0]
    
    def is_empty(self) -> bool:
        return len(self.__items) == 0
    
    def size(self) -> int:
        return len(self.__items)
    
    def print(self) ->int:
        print(f"{self.__items = }")
```

```python
queue_1: Queue = Queue()
queue_1.enqueue(1)
queue_1.enqueue(2)
queue_1.enqueue(3)
queue_1.enqueue(4)
print(f"{queue_1.dequeue() = }")
print(f"{queue_1.peek() = }")
queue_1.print()
queue_1.is_empty()
```

```python
# Usage expample
# https://runestone.academy/ns/books/published/pythonds3/BasicDS/SimulationHotPotato.html
def hot_potato(name_list, num):
    sim_queue = Queue()
    for name in name_list:
        sim_queue.enqueue(name)

    while sim_queue.size() > 1:
        for i in range(num):
            sim_queue.enqueue(sim_queue.dequeue())

        sim_queue.dequeue()
        sim_queue.print()

    return sim_queue.dequeue()


print(hot_potato(["Bill", "David", "Susan", "Jane", "Kent", "Brad"], 7))
```
