## Heavy function

```python
import math
import random
import time
from functools import lru_cache


def heavy_function(n: int = 5000) -> float:
    """
    A heavy function for profiling tests.
    Combines math, recursion, sorting, and simulated I/O.
    """
    # --- CPU-bound: Prime number generation ---
    def is_prime(num: int) -> bool:
        if num < 2:
            return False
        for i in range(2, int(math.sqrt(num)) + 1):
            if num % i == 0:
                return False
        return True

    primes = [x for x in range(2, n) if is_prime(x)]

    # --- Memory-bound: Sorting a large randomized dataset ---
    data = [random.random() * random.randint(1, n) for _ in range(n * 50)]
    data.sort()

    # --- Recursive + memoized Fibonacci ---
    @lru_cache(maxsize=None)
    def fib(x: int) -> int:
        if x < 2:
            return x
        return fib(x - 1) + fib(x - 2)

    fib_result = fib(25)  # intentionally small to avoid infinite recursion

    # --- Simulated I/O-like delay ---
    total_sleep = 0
    for _ in range(3):
        delay = random.random() * 0.1
        time.sleep(delay)
        total_sleep += delay

    # --- Numerical integration workload ---
    result = 0.0
    for i in range(1, n):
        x = i / n
        result += math.sin(x ** 2) * math.exp(-x)

    return result + sum(primes) + fib_result + total_sleep
```
