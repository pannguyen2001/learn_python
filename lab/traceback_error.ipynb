```python
from string import Template
error_message_template = Template("""Error in ${function_name}:
${traceback}""")
```

```python
import traceback

def test(first_num: int = 0, second_num: int = 0):
    try:
        return first_num / second_num
    except Exception as e:
        tb_str = "".join(traceback.TracebackException.from_exception(e).format())
        print(error_message_template.safe_substitute(function_name=__name__, error=e, traceback=tb_str))

def retest():
    return test()

def check(first_num: int = 0, second_num: int = 0) -> int:
    print("Check function")
    return retest()

check()
```
