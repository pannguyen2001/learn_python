```python
import random
import string

# Checksum tables
ST_CHECKSUM = "JZIHGFEDCBA"
FG_CHECKSUM = "XWUTRQPNMLK"

WEIGHTS = [2, 7, 6, 5, 4, 3, 2]


def calculate_checksum(prefix: str, digits: str) -> str:
    total = sum(int(d) * w for d, w in zip(digits, WEIGHTS))

    if prefix in ("T", "G"):
        total += 4

    if prefix in ("S", "T"):
        return ST_CHECKSUM[total % 11]
    elif prefix in ("F", "G", "M"):
        return FG_CHECKSUM[total % 11]
    else:
        raise ValueError("Invalid prefix")


def generate_nric(prefix: str = "S") -> str:
    if prefix not in ("S", "T", "F", "G", "M"):
        raise ValueError("Invalid NRIC/FIN prefix")

    digits = "".join(str(random.randint(0, 9)) for _ in range(7))
    checksum = calculate_checksum(prefix, digits)
    return f"{prefix}{digits}{checksum}"


def validate_nric(nric: str) -> bool:
    if len(nric) != 9:
        return False

    prefix = nric[0]
    digits = nric[1:8]
    checksum = nric[-1]

    if not digits.isdigit():
        return False

    try:
        return calculate_checksum(prefix, digits) == checksum
    except ValueError:
        return False
```
