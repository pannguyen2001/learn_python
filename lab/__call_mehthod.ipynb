```python
class LasWordInList:

    def __init__(self, input_list: list[str] = None) -> None:
        self.input_list = input_list

    def __call__(self) -> list:
        print(f"{self.__class__.__name__} called with input_list: {self.input_list}")
        return [_[-1] for _ in self.input_list]

if __name__ == '__main__':
    print(LasWordInList(['hello', 'world', 'python'])())
    print(LasWordInList(['hello', 'world', 'python', 'is', 'awesome'])())
```

```python
# factorial.py

class Factorial:
    def __init__(self):
        self.cache = {0: 1, 1: 1}

    def __call__(self, number):
        if number not in self.cache:
            # self(number - 1) == self.__call__(number - 1)
            self.cache[number] = number * self(number - 1)
            print(f"{number = }, {self.cache[number]= }")
        return self.cache[number]

if __name__ == '__main__':
    fact = Factorial()
    print(f"{fact(5) = }")
    # print(f"{fact(6) = }")
```

```python
class Logger:
    def __init__(self, filename):
        self.filename = filename

    def __call__(self, message):
        with open(self.filename, mode="a", encoding="utf-8") as log_file:
            log_file.write(message + "\n")

logger: Logger = Logger("test __call__ method in py.log")
logger("Hello world")
print("Complete logger.")
# You can apply in data validation or data processing class, no need write function like data_validation.run(), just implement __call__()
```
