## DSA basic - Stack
- Implement stack by using class and list

```python
class Stack:
    """Stack implemented by using list"""
    def __init__(self) -> None:
        self.__items: list = []
    
    def is_empty(self) -> bool:
        """Check if stack is empty"""
        return len(self.__items) == 0
    
    def push(self, item: any) -> list:
        """Add item to stack"""
        self.__items.append(item) # It return None
        print(f"After push: {self.__items = }")
        return self.__items
    
    def pop(self) -> None:
        """Remove item from stack"""
        pop_element: any = self.__items.pop()
        print(f"After pop: {self.__items = }")
        return pop_element
    
    def peek(self, item: any) -> list:
        """Get the top item from stack"""
        return self.__items[-1]

    def size(self) -> int:
        """Get size of stack"""
        return len(self.__items)
    
    def print_stack(self) -> None:
        print(f"Print stack: {self.__items = }")
```

```python
stack_1: Stack = Stack()
stack_1.push(1)
stack_1.push(2)
stack_1.push(3)
```

```python
m = Stack()
m.push("x")
m.push("y")
m.push("z")

step: int = 0
while not m.is_empty():
   step += 1
   print(f"Pop loop: {step}")
   m.print_stack()
   m.pop()
   m.print_stack()
   m.pop()
   m.print_stack()
```

```python
# General Balanced Symbol Problem
def par_checker(symbol_string):
    s = Stack()
    for symbol in symbol_string:
        if symbol == "(":
            s.push(symbol)
        else:
            if s.is_empty():
                return False
            else:
                s.pop()

    return s.is_empty()


print(par_checker("((()))"))  # expected True
print(par_checker("((()()))"))  # expected True
print(par_checker("(()"))  # expected False
print(par_checker(")("))  # expected False
```

```python
def balance_checker(symbol_string):
    s = Stack()
    for symbol in symbol_string:
        if symbol in "([{":
            s.push(symbol)
        else:
            if s.is_empty():
                return False
            else:
                if not matches(s.pop(), symbol):
                    return False

    return s.is_empty()

def matches(sym_left, sym_right):
    all_lefts = "([{"
    all_rights = ")]}"
    return all_lefts.index(sym_left) == all_rights.index(sym_right)


print(balance_checker('{({([][])}())}'))
print(balance_checker('[{()]'))
```

```python
# go deeper
def matches(sym_left, sym_right):
    all_lefts = "([{"
    all_rights = ")]}"
    print(f"{all_lefts.index(sym_left) = }, {all_rights.index(sym_right) = }")
    return all_lefts.index(sym_left) == all_rights.index(sym_right)

matches('(', ')')
matches('[', ']')
matches('{', '}')
matches('{', ']')
```
