## linked list

```python
class Node:
    def __init__(self, value: str|int|bool, pointer: "Node" = None) -> None:
        self.value = value
        self.pointer = pointer

# Node 1
head = Node(1)
node_2 = Node(2)
node_3 = Node(3)

# link node
head.pointer = node_2
node_2.pointer = node_3

```

```python
def print_linked_list(head:  Node = None) -> None:
    current = head
    while current:
        print(current.value)
        current = current.pointer
    print(None)
```

```python
print_linked_list(head)
```

```python
def add_node_at_first(head: Node = None, new_node: Node = None):
    new_node.pointer = head
    head = new_node
    return head
```

```python
head = add_node_at_first(head, Node(0))
```

```python
print_linked_list(head)
```

```python
def add_node_to_last(head: Node = None, new_node: Node = None) -> Node:
    if head is None:
        head = new_node
    
    last: Node = head
    while last.pointer is not None:
        last = last.pointer
    
    last.pointer = new_node
    return head
```

```python
head = add_node_to_last(head, Node(4))
print_linked_list(head)
```

```python
def add_node_to_specific_position(head: Node = Node, new_node: Node = None, position: int = 0):
    if not new_node:
        return None
    if not head:
        head = new_node
        return head
    if position == 0:
        add_node_at_first(head, new_node)
        return head
```

```python
# Complete

class Node:
    def __init__(self, value: str|int|bool, pointer: "Node" = None) -> None:
        self.value = value
        self.pointer = pointer

class LinkedList:
    def __init__(self, node: Node = None):
        self.head = node
    
    def print_linked_list(self) -> None:
        current = self.head
        while current:
            print(current.value)
            current = current.pointer
        print(None)
        
    def add_node_to_first(self, node: Node):
        node.pointer = self.head
        self.head = node
        return self.head
    
    def add_node_to_last(self, new_node: Node = None) -> Node:
        if self.head is None:
            self.head = new_node
        
        last: Node = self.head
        while last.pointer is not None:
            last = last.pointer
        
        last.pointer = new_node
        return self.head
```

```python
linked_list = LinkedList(Node(0))
linked_list.print_linked_list()
```

```python
linked_list.add_node_to_first(Node(-1))
linked_list.add_node_to_first(Node(-2))
linked_list.print_linked_list()
```

```python
linked_list.add_node_to_last(Node(1))
linked_list.add_node_to_last(Node(2))
linked_list.print_linked_list()
```
