```python
import string
import random

def generate_acra_uen() -> str:
    """
    Generate a random ACRA UEN number
    """
    uen = "".join(random.choices(string.digits, k=9)) + "".join(random.choices(string.ascii_uppercase, k=1))
    return uen

def generate_uen(uen_type: str = "ACRA") -> str:
    if uen_type == "ACRA":
        return generate_acra_uen()
    else:
        raise ValueError("Invalid UEN type")
```
