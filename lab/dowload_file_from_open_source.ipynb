```python
from huggingface_hub import hf_hub_download
import pandas as pd
import polars as pl

# Download a specific file to a local directory
# Goto https://huggingface.co/datasets/tmquan/sapnhap-bando-vn/edit/main/README.md and copy data_files path
file_path = hf_hub_download(
    repo_id="tmquan/sapnhap-bando-vn", 
    filename="data/all.parquet", 
    repo_type="dataset"
)

pl.scan_parquet(file_path).collect().write_parquet("./data/sapnhap-bando-vn.parquet")
```
