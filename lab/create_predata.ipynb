```python
import pandas as pd
from loguru import logger
from typing import Callable
from helpers.read_file import read_excel_file
from helpers.logger_wrapper import logger_wrapper

@logger_wrapper
def create_predata(file_path: str = "", sheet_name: str = "", create_data_func: Callable = None) -> pd.DataFrame|None:
    df = read_excel_file(file_path, sheet_name)
    if df is not None and not df.empty:
        logger.info(f"Predata retrieved successfully: {sheet_name}")
        return df

    logger.info(f"Predata is empty: {sheet_name}. Starting to create predata.")
    df = create_data_func(file_path, sheet_name)
    if df is not None and not df.empty:
        logger.info(f"Predata created successfully: {sheet_name}")
        return df

    logger.error(f"Error in retreiving/creating predata: {sheet_name}")
    return None
```
